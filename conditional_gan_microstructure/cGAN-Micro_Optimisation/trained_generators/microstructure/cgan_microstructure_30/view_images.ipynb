```python
import matplotlib.pyplot as plt
import tifffile as tiff
import numpy as np
import pandas as pd
```

```python
PATH = '/home/katiez/ML/conditional_gan_microstructure/cGAN-Micro_Optimisation/trained_generators/microstructure/cgan_microstructure_30'
numbers = np.around(np.linspace(0,1,9),2)
```

```python
f, axarr = plt.subplots(3,3)
f.set_figheight(10)
f.set_figwidth(10)
for n, ax in zip(numbers, axarr.ravel()):
    im = tiff.imread(f'cgan_microstructure_30[{str(n)}].tif')
    ax.imshow(im,cmap='gray')
    ax.set_title(str(n))
plt.tight_layout()
#plt.savefig('egg_raw_seed16')
plt.show()
```

```python
numbers = [-1,-2,-3,-4,-5,1.5,2,3,4,5]
f, axarr = plt.subplots(2,5)
f.set_figheight(10)
f.set_figwidth(10)
for n, ax in zip(numbers, axarr.ravel()):
    im = tiff.imread(f'cgan_microstructure_30[{str(n)}].tif')
    ax.imshow(im,cmap='gray')
    ax.set_title(str(n))
plt.tight_layout()
#plt.savefig('egg_raw_seed16')
plt.show()
```

```python
f, axarr = plt.subplots(3,3)
f.set_figheight(10)
f.set_figwidth(10)
for n, ax in zip(numbers, axarr.ravel()):
    df = pd.read_csv(PATH+f'/[{n}]/[{n}].csv')
    im = tiff.imread(f'cgan_microstructure_24[{str(n)}].tif')
    ax.imshow(im,cmap='gray')
    ax.plot(df['x'], df['y'], linestyle='', marker='o')
    #ax.set_ylim(0,40)
    ax.set_title(str(n))
plt.tight_layout()
#plt.savefig('mixture_measure_seed16')
plt.show()
```
